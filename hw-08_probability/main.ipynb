Для $U[0, 1]$ сгенерируйте $200$ независимых случайных величин, постройте график плотности на отрезке $[-0.25, 1.25]$, а также гистограмму по сгенерированным случайным величинам. Для построения графика необходимо брать достаточно большое количество точек сетки.

```python
def coin(n):
    return sps.bernoulli.rvs(p=0.5, size=n) #для нормальной работы используем функцию

def uniform(size=1, precision=30):
	matrix1 = coin(np.append(size, precision))
	matrix2 = np.logspace(1, precision, num=precision, base=.5)
	return (matrix1 @ matrix2)

def normal(size=1, loc=0, scale=1, precision=30):
	u1 = uniform(size, precision)
	u2 = uniform(size, precision)
	z = (-2 * np.log(u1))**0.5 * np.cos(2 * np.pi * u2)
	return loc + z * scale

def expon(size=1, lambd=1, precision=30):
	u = uniform(size, precision)
	return -(np.log(u) / lambd)
```

Исследуйте, как меняются значения случайных величин в зависимости от `precision` в написанной ранее вами функции.

Другой способ &mdash; смоделировать большое количество выборок.
Возьмем в качестве меры отклонения от среднего *среднее значение модуля отклонения от среднего*, то есть
$$F(n) = \frac{1}{K}\sum\limits_{j=1}^{K} \left| S_{nj} - a \right|,$$
где $a$ &mdash; математическое ожидание элемента выборки, $S_{nj}$ &mdash; *выборочное* среднее $n$ элементов в $j$-й выборке, а $K$ &mdash; общее количество выборок.

**1.** Рассмотрим распределения $Bern(a)$ для $a=0.5$. Пусть для нас допустимо отклонение на 0.02, иначе говоря при *истинном* среднем 0.5 мы считаем достаточно точными значения *выборочного* среднего от 0.48 до 0.52. Тогда в качестве *приближения* необходимого размера выборок можно взять наименьшее значение $n$, при котором $F(n) \leqslant 0.02$.

Реализуйте данный эксперимен для $K=500$  и найдите значение $n$. Визуализируйте зависимость $F(n)$ от $n$ в обычном и логарифмическом масштабах (по тем осям, по которым это имеет смысл).

Для реализации без циклов стоит генерировать достаточно большие выборки и использовать функцию `np.where`.

```python
max_n = 10**100

size = 10000  # количество случайных величин берем достаточно большое, чтобы точно не пропустить n
samples_count = 500 # K=500
samples = sps.bernoulli(p=0.5).rvs(size=(samples_count, size))
cum_means = samples.cumsum(axis=1) / (np.arange(size) + 1)
f = np.sum(abs(cum_means - 0.5), axis=0) / samples_count
n = np.arange(size) + 1
sentinel = size + 1
candidate = np.where(f <= 0.02, n, sentinel)
print(candidate.min())
```

```python
n_05 = []
for i in range(10):
    samples = sps.bernoulli(p=0.5).rvs(size=(samples_count, size))
    cum_means = samples.cumsum(axis=1) / (np.arange(size) + 1)
    f = np.sum(abs(cum_means - 0.5), axis=0) / samples_count
    n = np.arange(size) + 1
    sentinel = size + 1
    candidate = np.where(f <= 0.02, n, sentinel)
    result_n = candidate.min()
    n_05.append(result_n)      # <<< добавляем сюда
print(np.round(np.mean(n_05)))
```

```python
plt.figure(figsize=(8, 6))

sns.lineplot(x=n, y=f)
plt.xlabel("размер выборки n")
plt.ylabel("F(n)")
plt.title("Зависимость F(n) от минимального размера выборки")
#plt.xlim((0, 2000))
#plt.ylim((0, 0.15))
```

```python
plt.figure(figsize=(8, 6))

sns.lineplot(x=np.log(n), y=np.log(f))
plt.xlabel("ln(n)")
plt.ylabel("ln(F(n))")
plt.title("Зависимость ln(F(n)) от минимального размера выборки ln(n)")
```

#### Анализ графика:

График представлен в логарифмическом масштабе по обеим осям и демонстрирует линейную зависимость величины $F(n)$ от размера выборки $n$, что указывает на степенную зависимость вида $F(n) \sim n^q$, где $q$ — наклон прямой на логарифмическом графике. Значение $q < 0$ подтверждает гипотезу о том, что для повышения точности оценки среднего значения необходимо использовать выборки большего объема. При переходе к обычному масштабу наблюдается, что при увеличении размера выборки величина $F(n)$ изменяется всё менее существенно, что свидетельствует о сходимости соответствующего статистического оценивания.


**2.** Повторите данный эксперимент для значений $a$ от 0.05 до 0.95 с шагом 0.05. Визуализируйте зависимость необходимого размера выборки $n$ от значения $a$.

При реализации можно сделать цикл по значениям $a$.

```python
max_n = size + 1

min_n = []
a_quantity = []

a = 0.05
while a <= 0.95:
    samples = sps.bernoulli(p=a).rvs(size=(samples_count, size))
    cum_means = samples.cumsum(axis=1) / (np.arange(size) + 1)
    f = np.abs(cum_means - a).sum(axis=0) / samples_count

    n = np.arange(1, size + 1)
    mask = (f <= 0.02)
    if mask.any():
        result_n = mask.argmax() + 1
        min_n.append(result_n)
        a_quantity.append(a)
    a += 0.05
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(x=a_quantity, y=min_n)
plt.xlabel("a")
plt.ylabel("n")
plt.title("Зависимость мин.размера выборки n от параметра a")
```

Значение $q < 0$ подтверждает гипотезу о том, что для повышения точности оценки среднего значения необходимо использовать выборки большего объема. При переходе к обычному масштабу наблюдается, что при увеличении размера выборки величина $F(n)$ изменяется всё менее существенно, что свидетельствует о сходимости соответствующего статистического оценивания.

Это обусловлено тем, что функция отклонения от среднего значения, обозначаемая как $F(n)$, убывает по степенному закону $F(n) \sim n^{-q}$, где $q > 0$. Рассмотрим следующее вероятностное пространство:
* $\Omega = \{1, 2, 3, 4, 5, 6, 7, 8\}$ – номера точек на единичной окружности.

* $\sigma$-алгебра $\mathscr{F}$, которая включает все подмножества множества $\Omega$, $\mathscr{F} = 2^{\{1, 2, 3, 4, 5, 6, 7, 8\}}$
* $\mathsf{P}$ - равномерное распределение:
$$
\mathsf{P}(\omega = \{k\}) = \frac{1}{8}, \quad k = 1, 2, \ldots, 8.
$$

Определим случайную величину следующим образом:
$$
X(\omega) = \sin\left(\frac{2\pi (\omega - 1)}{8}\right) = \sin\left(\frac{\pi (\omega - 1)}{4}\right)$$
Интерпретация: значение координаты $y$ для точки с номером $w$ на единичной окружности.


Пусть случайная величина $Y$ определяется следующим образом:
$$ Y(\omega) = \begin{cases}
1, & \text{если } \sin\left(\frac{\pi (\omega - 1)}{4}\right) \geq 0, \\
0, & \text{иначе.}
\end{cases} $$


Какому наблюдению соответствует случайная величина $Y$?

$Y$ — это индикатор на единичной окружности относительно горизонтальной оси.

$$
Y(\omega)=
\begin{cases}
1,& \text{если точка с номером }\omega\text{ лежит на верхней полуокружности, то есть }y=\sin\theta\ge 0,\\[4pt]
0,& \text{если точка лежит на нижней полуокружности, то есть }y=\sin\theta<0.
\end{cases}
$$

Таким образом, наблюдение $Y$ фиксирует **знак $y$-координаты** выбранной точки:
\* $Y=1$ — «точка не ниже оси $Ox$»;
\* $Y=0$ — «точка ниже оси $Ox$».




Найдите:
* Условное распределение $X|Y$;
* Условное математическое ожидание $\mathsf{E}(X|Y=1)$;
* Условное математическое ожидание $\mathsf{E}(X|Y=0)$.

Визуализируйте условное распределение $X|Y$.



## 1 Модель

### 1.1 Исходное пространство

$$
\Omega=\{1,\dots ,8\}, \qquad
\mathbb P(\{\omega\})=\tfrac18\;\;(\omega\in\Omega).
$$

### 1.2 Случайные величины

$$
X(\omega)=\sin\!\Bigl(\tfrac{\pi(\omega-1)}{4}\Bigr),  
\qquad
Y(\omega)=\mathbf 1_{\{X(\omega)\ge0\}}.
$$



## 2 Условное распределение $X\mid Y$
| Y | Возможные $x$ | $P\bigl(X=x \mid Y=y\bigr)$ |
|:-:|:----------------|:------------------------------|
| 1 | $0$           | $2/5$                       |
|   | $\tfrac{\sqrt2}{2}$ | $2/5$                 |
|   | $1$           | $1/5$                       |
| 0 | $-\tfrac{\sqrt2}{2}$ | $2/3$                |
|   | $-1$          | $1/3$                       |


## 3 Условные математические ожидания

$$
\begin{aligned}
\mathbb E[X\mid Y=1] 
&= 0\cdot\frac25 
  +\frac{\sqrt2}{2}\cdot\frac25
  +1\cdot\frac15
  = \frac{\sqrt2+1}{5},\\[6pt]
\mathbb E[X\mid Y=0] 
&= \Bigl(-\frac{\sqrt2}{2}\Bigr)\!\cdot\frac23
  +(-1)\cdot\frac13
  = -\frac{\sqrt2+1}{3}.
\end{aligned}
$$



```python
omega = np.arange(1, 9)
x_vals = np.sin(np.pi * (omega - 1) / 4)
y_vals = (x_vals >= 0).astype(int)
df = pd.DataFrame({"ω": omega, "X": x_vals, "Y": y_vals})

def conditional_pmf(frame, y_value):
    mask = frame["Y"] == y_value
    return frame.loc[mask, "X"].value_counts(normalize=True).sort_index()

for y_val, title in [(1, "Y = 1 (верхняя)"), (0, "Y = 0 (нижняя)")]:
    pmf = conditional_pmf(df, y_val)
    ax = pmf.plot(kind="bar", figsize=(6,4))
    ax.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))
    for lbl in ax.get_xticklabels():
        lbl.set_rotation(45)
        lbl.set_ha("right")
    ax.set_title(f"Условное распределение X | {title}")
    ax.set_xlabel("x")
    ax.set_ylabel(f"P(X = x | Y = {y_val})")
    plt.tight_layout()
    plt.show()
```
